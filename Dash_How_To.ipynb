# Plot Dash

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
import numpy as np

# Generate synthetic condo data
np.random.seed(42)
num_data_points = 100
condos = pd.DataFrame({
    'DEVELOPMENT_jittered': np.random.uniform(0, 100, num_data_points),
    'Sold_Price': np.random.uniform(100000, 500000, num_data_points),
    'Bedrooms': np.random.choice([1, 2, 3], num_data_points)
})

# Create a Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    html.H1("Condo Data Analysis"),

    # Scatter plot
    dcc.Graph(id='scatter-plot')
])

# Callback to update the scatter plot
@app.callback(
    Output('scatter-plot', 'figure'),
    Input('scatter-plot', 'relayoutData')
)
def update_scatter_plot(relayoutData):
    # Create the scatter plot using Plotly Express
    scatter_fig = px.scatter(
        condos, x='DEVELOPMENT_jittered', y='Sold_Price', color='Bedrooms',
        title='Condo Data Scatter Plot', color_discrete_map={1: 'blue', 2: 'green', 3: 'red'}
    )

    return scatter_fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
import numpy as np

# Generate synthetic condo data
np.random.seed(42)
num_data_points = 100
condos = pd.DataFrame({
    'DEVELOPMENT_jittered': np.random.uniform(0, 100, num_data_points),
    'Sold_Price': np.random.uniform(100000, 500000, num_data_points),
    'Bedrooms': np.random.choice([1, 2, 3], num_data_points)
})

# Create a Dash app
app = dash.Dash(__name__)

# Layout of the app with two graphs side by side
app.layout = html.Div([
    html.H1("Condo Data Analysis"),

    # First Scatter plot
    html.Div([
        dcc.Graph(id='scatter-plot-1', style={'width': '49%', 'display': 'inline-block'}),

        # Second Scatter plot
        dcc.Graph(id='scatter-plot-2', style={'width': '49%', 'display': 'inline-block'}),
    ]),
])

# Callback to update the first scatter plot
@app.callback(
    Output('scatter-plot-1', 'figure'),
    Input('scatter-plot-1', 'relayoutData')
)
def update_scatter_plot_1(relayoutData):
    # Create the first scatter plot using Plotly Express
    scatter_fig_1 = px.scatter(
        condos, x='DEVELOPMENT_jittered', y='Sold_Price', color='Bedrooms',
        title='Condo Data Scatter Plot 1', color_discrete_map={1: 'blue', 2: 'green', 3: 'red'}
    )

    return scatter_fig_1

# Callback to update the second scatter plot
@app.callback(
    Output('scatter-plot-2', 'figure'),
    Input('scatter-plot-2', 'relayoutData')
)
def update_scatter_plot_2(relayoutData):
    # Create the second scatter plot using Plotly Express
    scatter_fig_2 = px.scatter(
        condos, x='Bedrooms', y='Sold_Price', color='Bedrooms',
        title='Condo Data Scatter Plot 2', color_discrete_map={1: 'blue', 2: 'green', 3: 'red'}
    )

    return scatter_fig_2

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
